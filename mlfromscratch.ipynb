```python
import pandas as pd
from sklearn.linear_model import LinearRegression

dane = [{"x" :[-1, -1, 0, 1, 3], "y" : [1, 1, 3, 4, 5]}]
data = pd.DataFrame({"x" :[-1, -1, 0, 1, 3], "y" : [1, 1, 3, 4, 5]})
```

```python
from matplotlib import pyplot as plt

plt.scatter(data['x'], data['y'])
```

```python


def gradientDescent(dane, lr, ls):
    x = dane['x']
    y = dane['y']
    n = len(data['x'])
    a = 1
    b = 2

    MSEs = []

    fig = plt.figure()
    fig, ax = plt.subplots(nrows=1, ncols=2)

    ax[0].scatter(data['x'], data['y'])

    for j in range(ls):
        suma = 0
        for i in range(n):
            suma += (y[i] - (a * x[i] + b))**2
        MSE = suma/n

        MSEs.append(MSE)

        ax[0].plot(x, a * x + b, 'r')

        sumaA = 0
        sumaB = 0

        for i in range(n):
            sumaA += -x[i]*(y[i] - (a * x[i] + b))
            sumaB += -(y[i] - (a * x[i] + b))

        pochodnaA = (2/n) * sumaA
        pochodnaB = (2/n) * sumaB

        a = a - (lr * pochodnaA)
        b = b - (lr * pochodnaB)
    ax[1].plot(MSEs)
    fig.show()

gradientDescent(data, 0.1, 10)
```
